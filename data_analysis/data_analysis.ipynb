## Imports necessários

```python
import pandas as pd
```

## Função de Visualização

```python
def print_perfil_acertos(sucessos):
    print("Perfil dos Acertos:")

    print("\nMelhores Funções de ", end="")
    print(sucessos["Selecao"].value_counts(normalize=True) * 100)

    print("\nMelhores Funções de ", end="")
    print(sucessos["Selecao_Best"].value_counts(normalize=True) * 100)

    print("\nMelhores Funções de ", end="")
    print(sucessos["Crossover"].value_counts(normalize=True) * 100)

    print("\nRank de ", end="")
    print(sucessos["Mutacao"].value_counts(normalize=True) * 100)
```

## Seleção dos Sucessos baseado na função:  (-(x-3)^2)+10 usando FITNESS_MAX

```python
df = pd.read_csv("historico_testes_func_1_MAX.csv")

sucessos = df[df["Resultado"] == 10]

print(f"Total de combinações testadas: {len(df)}")
print(f"Total de Acertos: {len(sucessos)} ({len(sucessos) / len(df) * 100:.2f}%)")

print_perfil_acertos(sucessos)
```

## Seleção dos Sucessos baseado na função: x^2 usando FITNESS_MIN1

```python
df_min1 = pd.read_csv("historico_testes_func_2_MIN1.csv")

sucessos_min1 = df_min1[df_min1["Resultado"] == 0]

print(f"\nTotal de combinações testadas: {len(df_min1)}")
print(f"Total de Acertos: {len(sucessos_min1)} ({len(sucessos_min1) / len(df_min1) * 100:.2f}%)")

print_perfil_acertos(sucessos_min1)
```
